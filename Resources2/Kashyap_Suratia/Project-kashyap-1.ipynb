```python
#import the pandas library amd matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load the excel datasheet
input_file = pd.ExcelFile("copy-of-kashyap-energy-and-tech-sector.xlsx")
#get the file sheets names
read_sheets_name = input_file.sheet_names
#print the sheet names
read_sheets_name
```

```python
# Merged graph for all the sheets
#load all the sheets by index
df1 = pd.read_excel(input_file,sheet_name =0)
df2 = pd.read_excel(input_file,sheet_name =1)
df3 = pd.read_excel(input_file,sheet_name =2)
df4 = pd.read_excel(input_file,sheet_name =3)
df5 = pd.read_excel(input_file,sheet_name =4)
df6 = pd.read_excel(input_file,sheet_name =5)
df7 = pd.read_excel(input_file,sheet_name =6)
df8 = pd.read_excel(input_file,sheet_name =7)

#Plot all the dataframes

plt.plot(df1['Date'], df1['Close'])
plt.plot(df2['Date'], df2['Close'])
plt.plot(df3['Date'], df3['Close'])
plt.plot(df4['Date'], df4['Close'])
plt.plot(df5['Date'], df5['Close'])
plt.plot(df6['Date'], df6['Close'])
plt.plot(df7['Date'], df7['Close'])
plt.plot(df8['Date'], df8['Close'])

# declare the X-axis and y-axis values
plt.xlabel('Date')
plt.ylabel('Close')
plt.title("Merged Sheet graphs")
plt.show
```


expanded data visualization for the merged sheets
