```python
#import the pandas library amd matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load the excel datasheet
input_file = pd.ExcelFile("copy_of_kashyap_energy_and_tech_sector_1.xlsx")
```

# c1_vgt_ks_csv vs c1_spy_ks_csv

```python
#load the sheets by tab name
df1 = pd.read_excel(input_file,sheet_name ='c1_vgt_ks_csv')
df2 = pd.read_excel(input_file,sheet_name ='c1_spy_ks_csv')

#Compare minimum, maximum and mean performance

values1 = df1[['Close']] # get the close values for c1_vgt_ks_csv
values2 = df2[['Close']] # get the close values for c1_spy_ks_csv

print("The Minimum close value for c1_vgt_ks_csv is: " +str(values1.min())) # print the minimum close value
print("The Minimum close value for c1_spy_ks_csv is: " +str(values2.min())) # print the minimum close value

print("The Maximum close value for c1_vgt_ks_csv is: " +str(values1.max())) # print the Maximum close value
print("The Maximum close value for c1_spy_ks_csv is: " +str(values2.max())) # print the Maximum close value

print("The Average close value for c1_vgt_ks_csv is: " +str(values1.mean())) # print the Average close value
print("The Average close value for c1_spy_ks_csv is: " +str(values2.mean())) # print the Average close value

#Plot all the dataframes

plt.plot(df1['Date'], df1['Close'],label="c1_vgt_ks_csv")
plt.plot(df2['Date'], df2['Close'],label="c1_spy_ks_csv")


# declare the X-axis and y-axis values
plt.xlabel('Date')
plt.ylabel('Close')
plt.title("c1_vgt_ks_csv vs c1_spy_ks_csv")
plt.legend()
plt.show
```
